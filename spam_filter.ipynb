```python
def clean_non_alpha(review_text):
  return re.sub('[^a-zA-Z]',' ',review_text)
df['clean_mail'] = df['text'].apply(clean_non_alpha)
```

```python
def convert_lowercase(review_text):
  return str(review_text).lower()
df['clean_mail'] = df['clean_mail'].apply(convert_lowercase)
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
def clean_tokenization(review_text):
  return word_tokenize(review_text)
df['clean_mail'] = df['clean_mail'].apply(clean_tokenization)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df.head()
```

```python
def clean_stopwords(token):
  return [item for item in list(token) if item not in stop_words ]
df['clean_mail'] = df['clean_mail'].apply(clean_stopwords)
```

```python
df.head()
```

```python
df['clean_mail'][2]
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def clean_stem(token):
  return [stemmer.stem(i) for i in token]
df['clean_mail'] = df['clean_mail'].apply(clean_stem)
```

```python
df['clean_mail'][2]
```

```python
def clean_length(token):
  return [i for i in token if len(i) > 2]
df['clean_mail'] = df['clean_mail'].apply(clean_length)
```

```python
df['clean_mail'][2]
```

```python
def convert_to_string(listMail):
  return ' '.join(listMail)
df['clean_mail'] = df['clean_mail'].apply(convert_to_string)
```

```python
df['clean_mail'][2]
```

```python
SpamMails = df[df.label == 'spam']['clean_mail']
HamMails = df[df.label == 'ham']['clean_mail']
```

```python
def remove_subject(text):
  return ' '.join(i for i in text.split() if i!= 'subject')
SpamMails = SpamMails.apply(remove_subject)
HamMails = HamMails.apply(remove_subject)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
xx = cv.fit_transform(df['clean_mail']).toarray()
```

```python
yy = pd.get_dummies(df['label'])
```

```python
yy = yy.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xx, yy, test_size = 0.20, random_state = 0)
```

```python
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
```

```python
print("The accuracy of our spam classifier is ",acc)
```
